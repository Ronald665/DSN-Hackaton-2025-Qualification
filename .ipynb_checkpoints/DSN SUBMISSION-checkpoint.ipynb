# Data Cleaning

```python
# Cleans Engine data column 
def clean_Engine_data(df):
    # Clean fuel_type column
    df = df.copy()
    mode_fuel = df.loc[~df["fuel_type"].isin(["–", "not supported"]), "fuel_type"].mode()[0]
    df["fuel_type"] = df["fuel_type"].replace(["–", "not supported"], mode_fuel)    
    df['fuel_type'] = df['fuel_type'].fillna('electric')
    
    # Extract numeric features from engine description
    df["engine_size_L"] = df["engine"].str.extract(r'([\d.]+)L').astype(float)
    df["cylinders"] = (
        df["engine"].str.extract(r'(\d+)\s*Cylinder')
        .astype(float)
        .astype("Int64")  # keeps missing values
    )
    df["horsepower"] = df["engine"].str.extract(r'([\d.]+)HP').astype(float)
    
    return df
```

```python
# Cleans External color columns
def clean_colors(df, col="ext_col"):
    basic_colors = {
        'black': 'Black',
        'white': 'White',
        'gray': 'Gray',
        'grey': 'Gray',
        'silver': 'Silver',
        'red': 'Red',
        'blue': 'Blue'
    }
    
    def map_to_basic(color):
        if pd.isna(color):   # handle NaN safely
            return "Other"
        for key, val in basic_colors.items():
            if key in color:
                return val
        return "Other"
    
    # normalize text and apply mapping
    df = df.copy()
    df[col] = df[col].str.strip().str.lower()
    df[f"{col}"] = df[col].apply(map_to_basic)
    
    return df
```

```python
# Cleans Transmission data Column
def clean_transmission_data(df):
    df = df.copy()
    
    # Find mode of valid transmission values
    mode_transmission = df.loc[~df["transmission"].isin(["–", "F", "2", "SCHEDULED FOR OR IN PRODUCTION"]), "transmission"].mode()[0]

    # Replace unwanted values with the mode
    df["transmission"] = df["transmission"].replace(["–", "F", "2", "SCHEDULED FOR OR IN PRODUCTION"], mode_transmission)

    # Replace any remaining NaNs with the mode
    df["transmission"] = df["transmission"].fillna(mode_transmission)

    # Extract numeric transmission speeds (e.g., "6-Speed", "10 Speed", "8-SPEED")
    df["transmission_speed"] = (
        df["transmission"].str.extract(r'(\d+)[ -]?[Ss]peed')
        .astype(float)
        .astype("Int64")  # keeps missing values
    )
    
    # Extract transmission type
    def get_type(value):
        value = str(value).lower()
        if "a/t" in value or "automatic" in value:
            return "Automatic"
        elif "m/t" in value or "manual" in value:
            return "Manual"
        elif "cvt" in value:
            return "CVT"
        elif "dct" in value:
            return "DCT"
        elif "variable" in value:
            return "Variable"
        elif "single-speed" in value:
            return "Single-Speed Fixed Gear"
        return "Other"
    
    df["transmission_type"] = df["transmission"].apply(get_type)
    
    return df
```

```python
# Cleans Interior Colors Column
def clean_interior_colors(df, col="int_col"):
    basic_colors_int = {
            'black': 'Black',
            'beige': 'Beige',
            'gray': 'Gray',
            'grey': 'Gray',
            'brown': 'Brown',
            'white': 'White',
            'red': 'Red'
    }
    def map_to_basic(color):
        if pd.isna(color):
            return "Other"
        color = color.lower().strip()
        for key, val in basic_colors_int.items():
            if key in color:
                return val
        return "Other"
    
    df = df.copy()
    df[col] = df[col].str.strip().str.lower()
    df[f"{col}"] = df[col].apply(map_to_basic)
    return df
```

```python
def clean_brand(df):
    luxury = {
        "Bugatti": "Luxury",
        "Lamborghini": "Luxury",
        "Rolls-Royce": "Luxury",
        "Bentley": "Luxury",
        "McLaren": "Luxury",
        "Ferrari": "Luxury",
        "Aston": "Luxury"
    }

    premium = {
        "Rivian": "Premium",
        "Porsche": "Premium",
        "Lucid": "Premium",
        "Maserati": "Premium",
        "Tesla": "Premium",
        "Maybach": "Premium",
        "Genesis": "Premium",
        "Land": "Premium",
        "Alfa": "Premium",
        "RAM": "Premium",
        "Mercedes-Benz": "Premium"
    }

    average = {
        "Jaguar": "Average",
        "Cadillac": "Average",
        "BMW": "Average",
        "Chevrolet": "Average",
        "Polestar": "Average",
        "Audi": "Average",
        "Ford": "Average",
        "GMC": "Average",
        "Kia": "Average",
        "Karma": "Average",
        "Lexus": "Average",
        "smart": "Average",
        "Jeep": "Average",
        "INFINITI": "Average",
        "Dodge": "Average",
        "FIAT": "Average",
        "Volvo": "Average",
        "Lincoln": "Average",
        "Suzuki": "Average"
    }

    cheap = {
        "Buick": "Cheap",
        "Nissan": "Cheap",
        "Toyota": "Cheap",
        "Acura": "Cheap",
        "Saab": "Cheap",
        "Plymouth": "Cheap",
        "Chrysler": "Cheap",
        "Lotus": "Cheap",
        "Volkswagen": "Cheap",
        "Mercury": "Cheap",
        "Hyundai": "Cheap",
        "Mazda": "Cheap",
        "Honda": "Cheap",
        "Pontiac": "Cheap",
        "Subaru": "Cheap",
        "Mitsubishi": "Cheap",
        "Saturn": "Cheap",
        "MINI": "Cheap",
        "Hummer": "Cheap",
        "Scion": "Cheap"
    }

    # Merge all dicts
    brand_category = {**luxury, **premium, **average, **cheap}

    # Map to dataframe
    df["brand_category"] = df["brand"].map(brand_category).fillna("Other")

    return df
```

```python
def handle_missing_values(df):
    df['transmission_speed'] =  df['transmission_speed'].fillna(df['transmission_speed'].median())
    df['horsepower'] =  df['horsepower'].fillna(df['horsepower'].median())
    df['cylinders'] =  df['cylinders'].fillna(df['cylinders'].median())
    df['clean_title'] =  df['clean_title'].fillna('no')
    df['engine_size_L'] =  df['engine_size_L'].fillna(df['engine_size_L'].median())
    df['accident'] =  df['accident'].fillna(df['accident'].mode()[0])
    
    df = df.drop('model', axis = 1)
    return df
```

# Feature Selection
Performed ANOVA, MI and Feature importance using Extra tree classifier to determine the most important features:
## Result:

### 1️⃣ Strong Core Features (Keep)

milage – consistently top predictor across F-test, MI, and XGBoost

model_year – strong linear and nonlinear signal

horsepower – highly informative

engine_size_L – consistently useful

transmission_speed – moderately strong, contributes meaningful signal

### 2️⃣ Secondary / Useful Features (Optional)

int_col – moderately useful

ext_col – decent contribution

brand_category – modest signal

cylinders – low individual contribution, may overlap with horsepower/engine_size

clean_title – weak XGB signal but strong F-test; may help in edge cases

### 3️⃣ Weak / Drop 

fuel_type – low contribution, likely redundant

accident – dataset bias; not generalizable

transmission_type – negligible signal

## ✅ Recommendation:

after trial and error drooping the weak features and using both core and secondary predictors gave the best result

Keep Core Predictors: milage, model_year, horsepower, engine_size_L, transmission_speed

Try Secondary Predictors: int_col, ext_col, brand_category, cylinders, clean_title (evaluate performance with/without)

Deprioritize: fuel_type, accident, transmission_type
